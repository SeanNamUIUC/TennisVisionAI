```python
import cv2
print(cv2.version)

video_path = '/content/tennis.mp4'
cap = cv2.VideoCapture(video_path)#open video file -> 영상에서 프레임 하나씩 꺼내오기 가능해짐
print('영상 열림?', cap.isOpened())
```

```python
ret, frame = cap.read() #프레임 읽기 성공여부, 이미지 한장

print(ret)
print(frame.shape) #높이, 너비, RGB channel (프레임 이미지 크기)
```

```python
from matplotlib import pyplot as plt

frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.imshow(frame_rgb)
plt.axis("off")
plt.show()
```
